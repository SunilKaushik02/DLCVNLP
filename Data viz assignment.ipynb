```python
# load the dataset

iris = pd.concat([pd.DataFrame(datasets.load_iris().data,columns=datasets.load_iris().feature_names),pd.Series(datasets.load_iris().target,name='target')],axis=1)

# iris['target']=iris['target'].map({0:'setosa',1:'versicolor',2:'virginica'})

iris.head()
```

```python
# plotting

X_reduced = decomposition.PCA(n_components = 3).fit_transform(iris.iloc[:,0:3])

Y=iris.target

fig = plt.figure(1 , figsize = (12,10))
ax = Axes3D(fig)

ax.scatter(X_reduced[: , 0] , X_reduced[:, 1] , X_reduced[:, 2] ,c=Y,cmap='plasma',edgecolor='k',s=20)

ax.set_title('Scatterplot of iris dataset')
ax.set_xlabel('1st component')
ax.set_ylabel('2nd component')
ax.set_zlabel('3rd component')

ax.legend(labels=['Setosa','Versicolor','Virginica'])
```
